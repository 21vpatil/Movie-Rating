```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to the CSV file
csv_path = 'Movie_Rating.csv'

# Load the dataset
df = pd.read_csv(csv_path)

# Display the first few rows of the dataset
print(df.head())

# Convert 'Release Date' to datetime format
df['Release Date'] = pd.to_datetime(df['Release Date'])

# Step 1: Create a mapping of old genres to new genres
genre_mapping = {
    'Action': 'Adventure',
    'Drama': 'Romantic',
    'Comedy': 'Humor',
    'Thriller': 'Suspense',
    'Horror': 'Scary',
    # Add more genre mappings as needed
}

# Step 2: Replace genres in the dataset
df['Genre'] = df['Genre'].replace(genre_mapping)

# Basic EDA with updated genres
# 1. Distribution of Movie Genres
plt.figure(figsize=(10, 6))
sns.countplot(y='Genre', data=df, order=df['Genre'].value_counts().index)
plt.title('Distribution of Movie Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()

# 2. Distribution of Movie Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# 3. Trend of Movie Releases Over Time
plt.figure(figsize=(12, 6))
df['Year'] = df['Release Date'].dt.year
sns.countplot(x='Year', data=df, order=sorted(df['Year'].unique()))
plt.title('Trend of Movie Releases Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.xticks(rotation=90)
plt.show()

# 4. Movie Duration Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Duration'], bins=20, kde=True)
plt.title('Distribution of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.show()

# 5. Relationship Between Rating and Duration
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Duration', y='Rating', data=df)
plt.title('Relationship Between Movie Duration and Rating')
plt.xlabel('Duration (minutes)')
plt.ylabel('Rating')
plt.show()
```
